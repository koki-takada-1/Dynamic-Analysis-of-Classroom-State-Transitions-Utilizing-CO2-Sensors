```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# CSVファイルを読み込む
df = pd.read_csv("2024-0110-371data.csv")

# '_time'列を日付型に変換
df["_time"] = pd.to_datetime(df["_time"])

# '_time'をx軸に、'co2'をy軸に設定してグラフを描画
plt.figure(figsize=(10, 6))
plt.plot(df["_time"], df["co2"], marker="o")

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%H:%M"))  # x軸のラベルを時間のみに設定
plt.gca().xaxis.set_major_locator(mdates.HourLocator())  # x軸の目盛りを1時間ごとに設定

plt.xlabel("時間")
plt.ylabel("CO2")
plt.title("時間によるCO2の変化")
plt.grid(True)

plt.show()
```
